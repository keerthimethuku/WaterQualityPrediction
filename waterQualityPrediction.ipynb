```python
import pandas as pd

# Load dataset correctly
df = pd.read_csv('/content/dataSet.csv', sep=';')

# Drop non-useful columns
df.drop(['id', 'date'], axis=1, inplace=True)

# Convert all columns to numeric (if any string issues)
df = df.apply(pd.to_numeric, errors='coerce')

# Fill missing values with mean
df.fillna(df.mean(), inplace=True)

# Confirm no missing values
print(df.isnull().sum())

# Split features and target (example: predicting O2)
X = df.drop('O2', axis=1)
y = df['O2']
```

```python
# Create a new column 'Potability' based on simple thresholds
df['Potability'] = ((df['O2'] >= 6) & (df['NH4'] <= 0.5) & (df['NO3'] <= 10)).astype(int)

# Check value counts
print(df['Potability'].value_counts())
```
